```python
data = cvrp.read_vrp("lit_instances/vrpnc5.txt")
prob = cvrp.CVRPProblem(data)
routes = [clark_wright.Route([0, i, 0], prob.weights[i]) for i in range(1, prob.num_clients+1)]
```

```python
sol_best = clark_wright.cw_parallel(routes, prob.distance_matrix, prob.vehicle_capacity)
map(lambda x: clark_wright.steepest_improvement(x, prob.distance_matrix), sol_best)
benchmark.print_solution(prob.positions, prob.distance_matrix, prob.weights, sol_best)
```

```python
def monte_carlo(num, k):
    assert(num > 0)
    sols = []
    for _ in range(num):
        sol = clark_wright.cw_parallel_random(routes, prob.distance_matrix, prob.vehicle_capacity, k)
        map(lambda x: clark_wright.steepest_improvement(x, prob.distance_matrix), sol)
        dist = benchmark.get_distance(sol, prob.distance_matrix, prob.weights)
        sols.append((sol, dist))
    return sorted(sols, key=lambda x:x[1])

#benchmark.print_solution(prob.positions, prob.distance_matrix, prob.weights, sol)
```

```python
%time sols = monte_carlo(300, 5)
```
