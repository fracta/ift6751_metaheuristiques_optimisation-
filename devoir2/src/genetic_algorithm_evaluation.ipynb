# Genetic Algorithm Benchmark

The genetic algorithm is used to solve 7 problem instances from Christofides as described in the report.
Results are compared with best known solutions and are used to benchmark this algorithm against the Random Savings and Tabu Search algorithms.

```python
%pylab inline
import pyximport
pyximport.install(reload_support=True)
```

```python
import benchmark
import cvrp
import ga_solver
import timeit
```

```python
# Best known solutions to Christofide's instances (without distance limit)
best_known = benchmark.best_known
problem_names = sorted(best_known.keys())
```

```python
# Read the data and arrange by increasing number of clients
data = {name : cvrp.read_vrp("../lit_instances/{0}.txt".format(name)) for name in problem_names}
problems = [cvrp.CVRPProblem(d, name) for name, d in data.items()]
problems.sort(key=lambda x:x.num_clients)
```

```python
# solve with the genetic algorithm

# set the seed for numpy's pseudorandom number generator
# for this case, all other parameters should be fixed and since no other prng is used,
# this should be repeatable on other computer with same architecture
np.random.seed(42)

k = 10  # exploration depth for the random savings

n_replicates = 5  # number of replicates to run

population_size = 200  # population size

num_generations = 100  # num generations

recombination_prob = 0.5  # 

mutation_prob = 0.1  # 

# create the results dict and fill it
results = dict()
for problem in problems:
    results[problem.problem_name] = []


# run the iterations
for iteration in range(n_replicates):
    print "replicate {0}".format(iteration)
    for problem in problems:
        print "solving {0}".format(problem.problem_name)
        # time the process
        start = timeit.default_timer()
        solutions = ga_solver.solve(problem, population_size, num_generations, k=k)
        end = timeit.default_timer()
        diff = end - start
        # 
        distances = map(lambda x: sum(x.get_information(problem.distance_matrix, problem.weights)["dist"]), solutions)
        best_dist = np.min(distances)
        print best_dist
        sol = solutions[np.argmin(distances)]
        # 
        results[problem.problem_name].append((problem.problem_name, diff, sol, best_dist))
    print ""
```

```python
# let's see what the best solutions look like
best_solutions = []

for problem in problems:
    name = problem.problem_name
    distances = map(lambda x:x[3], results[name])
    (n, diff, sol, best_dist) = results[name][np.argmin(distances)]
    print "Problem {0}    Best known result {1}".format(name, best_known[name])
    benchmark.display(problem, sol)
    best_solutions.append((n, diff, sol, best_dist))
```

```python
seconds_to_mins = lambda seconds: divmod(seconds, 60)

relative_results = []
problem_avg_time = []

for problem in problems:
    optimal_value = best_known[problem.problem_name]
    tmp = []
    cumul_time = 0.
    for (name, time, sol, best_dist) in results[problem.problem_name]:
            cumul_time += time
            rel_dist = best_dist / optimal_value
            tmp.append(rel_dist)
    m, s= seconds_to_mins(cumul_time / len(results[problem.problem_name]))
    problem_avg_time.append("{}\n{} min {} sec".format(problem.problem_name, int(m), int(s)))
    relative_results.append(tmp)
print relative_results

print problem_avg_time
```

```python
# let's plot the result of each computation
import seaborn
f, ax = plt.subplots(figsize=(12, 9))
seaborn.heatmap(relative_results, annot=True, vmin=1., vmax=1.1, fmt=".4", xticklabels=np.arange(1,6), yticklabels=problem_avg_time)
plt.xlabel("replicate")
plt.ylabel("problem instance")
plt.title("Genetic algorithm best values divided by best known values")
```
