# Crossover illustration

```python
prob2 = cvrp.CVRPProblem(cvrp.read_vrp("../lit_instances/vrpnc3.txt"))
sols = clark_wright.random_savings(prob2, 20, 10)
```

```python

sol1 = sols[np.random.randint(0, len(sols)-1)]
sol2 = sols[np.random.randint(0, len(sols)-1)]

for route in sol1.routes:
    two_opt.steepest_improvement(route, prob2.distance_matrix)
for route in sol2.routes:
    two_opt.steepest_improvement(route, prob2.distance_matrix)
```

```python
indices = ga_solver.find_inherited_routes(sol1, sol2, prob2.positions)
inherited_routes = []
r1 = []
r2 = [] 
for index in indices[0]:
    inherited_routes.append(sol1.routes[index])
    r1.append(sol1.routes[index])
for index in indices[1]:
    inherited_routes.append(sol2.routes[index])
    r2.append(sol2.routes[index])
```

```python
# 
unserved_clients = ga_solver.find_unserved_clients(inherited_routes, prob2.num_clients)
new_routes = []
for client in np.arange(1, prob2.num_clients+1):
    if client in unserved_clients:
        new_routes.append(Route([0, client, 0], prob2.weights[client]))
    else:
        new_routes.append(None)
remaining_routes = clark_wright.cw_parallel(new_routes,
                                            prob2.distance_matrix,
                                            prob2.vehicle_capacity)
for route in remaining_routes:
    two_opt.steepest_improvement(route, prob2.distance_matrix)
inherited_routes.extend(remaining_routes)

sol3 = Solution(inherited_routes)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

benchmark.display_routes_col(prob2, sol1.routes, "red", ax=axes[0, 0], show_legend=False, figsize=(8,8), title="parent 1")
benchmark.display_routes_col(prob2, sol2.routes, "blue",ax=axes[1, 0],  show_legend=False, figsize=(8,8),  title="parent 2")


benchmark.display_routes_col(prob2, r1, "red", ax=axes[0, 1], show_legend=False, figsize=(8,8), title="")
benchmark.display_routes_col(prob2, r2, "blue",ax=axes[0, 1], show_legend=False, figsize=(8,8), title="inherited routes")


benchmark.display_routes_col(prob2, sol3.routes,"green",ax=axes[1,1], figsize=(8, 8), show_legend=False, title="")
benchmark.display_routes_col(prob2, r1, "red", ax=axes[1,1], show_legend=False, figsize=(8,8),  title="")
benchmark.display_routes_col(prob2, r2, "blue",ax=axes[1,1], show_legend=False, figsize=(8,8),  title="children")

savefig("../../presentation/figs/crossover.png")
```
