```python
import benchmark
import clark_wright
import cvrp
from routes import Route
```

```python
# Best known solutions to Christofide's instances (without distance limit)
best_known = benchmark.best_known
problem_names = sorted(best_known.keys())
```

```python
# Read the data and arrange by increasing number of clients
data = {name : cvrp.read_vrp("../lit_instances/{0}.txt".format(name)) for name in problem_names}
problems = [cvrp.CVRPProblem(d, name) for name, d in data.items()]
problems.sort(key=lambda x:x.num_clients)
```

# Random savings experimental results

```python
k = 3  # exploration depth for the random savings

time_limit = 30  # 30 seconds for each problem

random_savings_results = dict()

for problem in problems:
    print "solving {0}".format(problem.problem_name)
    solutions = clark_wright.random_savings(problem, time_limit, k, problem.distance_matrix, problem.vehicle_capacity)
    distances = map(lambda x: benchmark.get_distance(x, problem.distance_matrix, problem.weights), solutions)
    best_distance = np.min(distances)
    best_solution = solutions[np.argmin(distances)]
    random_savings_results[problem.problem_name] = (best_distance, best_solution)
```

```python
# let's see what the solutions look like
for problem in problems:
    name = problem.problem_name
    best_solution = random_savings_results[name][1]
    print "Problem {0}    Best known result {1}".format(name, best_known[name])
    benchmark.print_solution(problem.positions, problem.distance_matrix, problem.weights, best_solution)
```

# Analysis

Given that only 30 seconds are used on each problem, the results are impressive.

Most of the results are really close to the optimal value.

However, the gap to the best known value grows as the number of clients (and number of routes) grows. This makes sense given that the random savings approach relies on the ordering of savings and the probability of an optimal solution not being attainable this way might grow as the number of clients grows.

Obviously, the shape of the clusters of clients and the position of the depot relative to the clients is an important factor to consider.
